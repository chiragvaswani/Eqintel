## Numerical Feature Selection

```python
from sklearn.feature_selection import SelectKBest, f_regression
```

```python
# Using the correlation statistic written in f_regression
fs = SelectKBest(score_func=f_regression, k=16)
```

```python
X_train = services
y_train = services['ROE']
```

```python
X_train.drop(columns=['SECTOR', 'TIC', 'GVKEY', 'ROE'], inplace=True)
```

```python
fs.fit(X_train, y_train)
```

```python
X_train_fs = fs.transform(X_train)
```

```python
# what are scores for the features
services_scores = []
for i in range(len(fs.scores_)):
    services_scores.append(fs.scores_[i])
    print('Feature %d: %f' % (i, fs.scores_[i]))
# plot the scores
plt.bar([i for i in range(len(fs.scores_))], fs.scores_)
plt.show()
```

```python
X_train = technology
y_train = technology['ROE']
X_train.drop(columns=['SECTOR', 'TIC', 'GVKEY', 'ROE'], inplace=True)
```

```python
def get_feature_score(sector):
    X_train = sector
    y_train = sector['ROE']
    X_train.drop(columns=['SECTOR', 'TIC', 'GVKEY', 'ROE'], inplace=True)
    fs = SelectKBest(score_func=f_regression, k=16)
    fs.fit(X_train, y_train)
    X_train_fs = fs.transform(X_train)
    services_scores = []
    for i in range(len(fs.scores_)):
        services_scores.append(fs.scores_[i])
        print('Feature %d: %f' % (i, fs.scores_[i]))
    # plot the scores
    plt.bar([i for i in range(len(fs.scores_))], fs.scores_)
    plt.show()
    print(X_train.info())
    return fs.scores_
```

```python
get_feature_score(services)
```
